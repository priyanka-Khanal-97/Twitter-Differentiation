 <center> <h2> Importing Various Python Libraries </h2> </center>
 

```python
from twitter_keys import consumer_key, consumer_secret, access_token, access_secret
import tweepy #for twitter API
import pandas as pd # To Handle Data
import numpy as np #for number computing
#for visualization and plotting
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from collections import Counter
from IPython.display import Image
import pickle #for pickling
#sentiment Analysis
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

<center> <h2> Authentication </h2> </center>

```python
#twitter API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth)
#Twitter account1: FC Barcelona
FCBarcelona_tweets = api.user_timeline('FCBarcelona', count=40)
#Twitter account2: Elon Musk
elonmusk_tweets = api.user_timeline('elonmusk', count=40)
#Twitter account3: jackflims
jacksfilms_tweets = api.user_timeline('jacksfilms', count=40)
```

<center> <h2> FCBarcelona Tweets Analysis </h2> <center> <br>
<img src="FCBarcelona.jpg" width=200>
 Futball club Barcelona is a professional football club based in Barcelona, Spain. FC Barcelona joined twitter on December 2009. FC Barcelona currently has 29.1M followers.                                                    Twitter Handle: @FCBarcelona

```python
#pickling tweets of FCbarcelona
#with open('FCBarcelona.pkl', 'wb') as f:
    #pickle.dump(FCBarcelona_tweets, f)
with open('FCBarcelona.pkl', 'rb') as f:
    FCBarcelona_tweets = pickle.load(f)
data = pd.DataFrame(data=[tweet.text for tweet in FCBarcelona_tweets], columns=['FCBTweets'])
data['FCBlen']  = np.array([len(tweet.text) for tweet in FCBarcelona_tweets])
data['FCBDate'] = np.array([tweet.created_at for tweet in FCBarcelona_tweets])
data['FCBSource'] = np.array([tweet.source for tweet in FCBarcelona_tweets])
data['FCBLikes']  = np.array([tweet.favorite_count for tweet in FCBarcelona_tweets])
data['FCBRTs']    = np.array([tweet.retweet_count for tweet in FCBarcelona_tweets])
display(data.head(40))
FCBmeanLikes = np.mean(data['FCBLikes'])
FCBmeanLen = np.mean(data['FCBlen'])
FCBmeanRts=np.mean(data['FCBRTs'])
FCBlexemes = []
for tweet in FCBarcelona_tweets:
    FCBlexemes.extend(tweet.text.split())
def lexical_diversity(tokens):
    return len(set(tokens))/len(tokens)
FCBld=lexical_diversity(FCBlexemes)
FCBcommonwords=(Counter(FCBlexemes).most_common(10))
df_FCBarcelona = pd.DataFrame.from_dict(FCBcommonwords)
df_FCBarcelona=df_FCBarcelona.rename(columns={0:'FCBCommon words', 1:'FCBcount'})
#df_FCBarcelona
```

```python
ax=df_FCBarcelona.plot.bar(x='FCBCommon words', y='FCBcount', rot=0, figsize=(15,5))
```

<center> <h2> Sentiment Analysis </h2> </center>

```python
def plot_sent(tweets):
    texts=[tweet.text for tweet in tweets]
    sentiments=[analyzer.polarity_scores(t) for t in texts]
    pos_x=[s['pos'] for s in sentiments]
    neu_y=[s['neu'] for s in sentiments]
    neg_z=[s['neg'] for s in sentiments]
    return pos_x,neu_y, neg_z
```

```python
#Plotting in 3D
#pos_x=[sent['pos'] for sent in FCBarcelona_sentiments]
#neu_y=[sent['neu'] for sent in FCBarcelona_sentiments]
#neg_z=[sent['neg'] for sent in FCBarcelona_sentiments]
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection='3d')
fcbx,fcby,fcbz= plot_sent(FCBarcelona_tweets)
emx, emy, emz = plot_sent(elonmusk_tweets)
jfx, jfy, jfz = plot_sent(jacksfilms_tweets)
ax.scatter(fcbx,fcby,fcbz,c='r', marker='o', label="FCBarcelona")
ax.scatter(emx, emy, emz,c='y', marker='^', label="elonmusk")
ax.scatter(jfx, jfy, jfz,c='g', marker='x', label="jacksfilms")
ax.set_xlabel('Positive')
ax.set_ylabel('Neutral')
ax.set_zlabel('Negative')
leg = ax.legend();
plt.show()
```

<img src="sentAna.png" width=1000>
The above 3d plot visualizes if the tweets were postive, negative or neutral. Here the red circle represents FCBarcelona, yellow triangle represents elonmusk and green X represents jacksfilms. 
As we can see, most of the tweets for all three twitter handles are positive or neutral where as there as some outliers tweets of Elon musk and jackfilms to be negative or neutral. The reason FCBarcelona has most tweets that are positive because it is a fan oriented twitter page.
